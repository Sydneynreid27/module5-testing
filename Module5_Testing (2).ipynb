```python
## Testing Your Code

Testing is the process of checking whether the code works correctly.

There are two main types:

- Manual Testing: running the program and checking results yourself
- Automated Testing: writing code to test code automatically

Automated testing is more reliable because it reduces human error and can be repeated easily.
```

```python
## Unit Tests vs Integration Tests

- Unit tests check a single function or a small part of a program
- Integration tests check how multiple parts of a program work together

Unit testing helps identify problems quickly by isolating code.
```

```python
def sum_values(values):
    total = 0
    for v in values:
        total += v
    return total
```

```python
# Writing simple unit tests using assert

def test_sum():
    assert sum_values([1, 2, 3]) == 6

def test_sum_tuple():
    assert sum_values((1, 2, 2)) == 5

def test_sum_empty():
    assert sum_values([]) == 0

test_sum()
test_sum_tuple()
test_sum_empty()

print("All tests passed successfully!")
```

```python
## Executing Your First Test

The tests are executed by calling the functions directly.

If all assertions are correct:
- The program runs without errors
- "All tests passed successfully!" is displayed

If a test fails:
- Python raises an AssertionError
- The error shows what the expected vs actual values were
```
